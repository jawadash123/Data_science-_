```python
import time

# Function to display the Red light
def red_light():
    print("Traffic Light is Red. Stop!")

# Function to display the Yellow light
def yellow_light():
    print("Traffic Light is Yellow. Get Ready!")

# Function to display the Green light
def green_light():
    print("Traffic Light is Green. Go!")

# Function to manage the traffic light system
def traffic_light_system(current_state):
    if current_state == "red":
        red_light()
    elif current_state == "yellow":
        yellow_light()
    elif current_state == "green":
        green_light()
    else:
        print("Invalid state! Please enter red, yellow, or green.")

# Main function to control the traffic light with a loop
def main():
    states = ["red", "green", "yellow"]
    durations = {"red": 5, "green": 5, "yellow": 2}  # Set duration for each state
    current_index = 0
    
    while True:
        current_state = states[current_index]
        traffic_light_system(current_state)
        time.sleep(durations[current_state])  # Wait for the duration of the current state
        current_index = (current_index + 1) % len(states)  # Cycle through the states

# Execute the main function
main()
```
